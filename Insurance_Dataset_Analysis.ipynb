Q4. Identify numerical and categorical columns in the dataset. Print them separately.

```python
# ======= numerical columns in the datase ========

print("Numerical Columns:")
print(df.select_dtypes(include="number").columns)

# select_dtypes() selects columns based on their data type.
# include = 'number' selects all numerical columns (such as int and float).
# .columns displays only the names of those columns.

# ======== categorical columns in the datase =======

print("\nCategorical Columns:")
print(df.select_dtypes(include="object").columns)

# include='object' selects all text (string) columns.
```

Q5. Create distribution plots for all numerical columns (age, bmi, children, charges):
1) Use sns.histplot() or sns.distplot()
2) Create one plot per column (total 4 plots)
3) Add meaningful titles and labels

```python
# List of numerical columns
columns = ["age", "bmi", "children", "expenses"]

# Create one histogram for each column
for col in columns:
    sns.histplot(df[col])
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

Q6. Create count plots for all categorical columns (sex, smoker, region):
1) Use sns.countplot() for each column
2) Add titles and labels for clarity

```python
# List of categorical columns
columns = ["sex", "smoker", "region"]

# Create count plot for each categorical column
for i in columns:
  sns.countplot(x = df[i], data = df)
  plt.title(f"Count of {i}")
  plt.xlabel(i)
  plt.ylabel("Count")
  plt.show()
```

Q7. Create a heatmap to show the correlation between numerical variables.
1) Use sns.heatmap()
2) Annotate the values and use a suitable color palette

```python
# Calculate correlation of numerical columns
corr = df.corr(numeric_only = True)

# Create heatmap
sns.heatmap(corr, annot=True, cmap="Reds")

# Add title
plt.title("Correlation Heatmap")

# Display heatmap
plt.show()
```

Q8. Perform the following analysis on the charges column:
1) Find the average insurance charges
2) Find the maximum and minimum charges
3) Group by smoker and compare the average charges for smokers vs nonsmokers

```python
# Find the average insurance charges
print("1) Average Expenses:")
print(df["expenses"].mean())

# Find the maximum insurance expenses
print("\n2) Maxmimum Expenses:")
print(df["expenses"].max())

# Find the minimum insurance expenses
print("\n3) Minimum Expenses:")
print(df["expenses"].min())

# Compare average charges of smokers and non-smokers
print("\n4) Average Expenses for Smokers:\n")
print(df.groupby("smoker")["expenses"].mean())
```

Q9. Create a boxplot to show the distribution of charges with respect to smoker and sex.

```python
# Create a boxplot of expenses with respect to smoker and sex.
sns.boxplot(x = df["smoker"], y = df["expenses"], hue = df["sex"])
plt.title("Boxplot of Expenses with respect to Smoker and Sex")
plt.xlabel("Smoker")
plt.ylabel("Expenses")
plt.show()
```
