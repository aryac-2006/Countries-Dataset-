```python
df = pd.read_csv(filepath)
df
```

```python
# Display the first 10 rows

print("<<<----------------------------<< First 10 Rows >>----------------------------->>>\n")
df.head(10)
```

```python
# ===== Check missing values ======
print("<<<<<<<<<<<  Missing Values  >>>>>>>>>>\n")
print(df.isnull().sum())

# ====== Calculate missing value percentage =======
missing_per = (df.isnull().sum() / len(df)) * 100
print("\n>>>>>>>>>>>  Missing Value Percentage  <<<<<<<<<<<<\n")
print(missing_per)

# Handle missing values
# Drop rows containing missing values
df_cleaned = df.dropna()

# Check duplicate rows
print("\n------------- Number of duplicate rows ---------------")
print(df.duplicated().sum())

# Remove duplicate rows
df = df.drop_duplicates()

# Check final Dataset
print("\n:::::::::: Final Dataset Shape :::::::::::\n")
print(df.shape)
```

```python
# Statistical summary of numeric columns
print("<<<<<<<<<<<<<< Statistical summary >>>>>>>>>>>>>\n")
print(df.describe())

# Target variable
target = "Price"

# Find min, max, mean and median
print("\n>>>>>>>>>>>> Target variable statistics <<<<<<<<<<<\n")
print("1) Minimum Price:-", df[target].min())
print("\n2) Maximum Price:-", df[target].max())
print("\n3) Mean Price:-", df[target].mean())
print("\n4) Median Price:-", df[target].median())
```

```python
# Select all numerical columns
numeric_col = df.select_dtypes(include = ['int64', 'float64']).columns

plt.figure(figsize = (15, 16))
for i, col in enumerate(numeric_col, 1):
  plt.subplot(4, 2, i)
  sns.histplot(df[col], kde = True)
  plt.title(f"Distribution of {col}")
  plt.xlabel(col)
  plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# Find categorical columns
categorical_col = df.select_dtypes(include = ['object']).columns
print("---------------------<< Categorical columns >>------------------\n")
print(categorical_col)

plt.figure(figsize = (13, 9))

# Create count plots
print("\n-------------------------<< Count plots >>----------------------\n")
for i , col in enumerate(categorical_col, 1):
  plt.subplot(2, 2, i)
  sns.countplot(data = df, x = col)
  plt.title(f"Count Plot of {col}")
  plt.xlabel(col)
  plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
# Select numeric columns
numeric_col = df.select_dtypes(include = ['int64', 'float64']).columns

# Create correlation heatmap
plt.figure(figsize = (10, 6))

sns.heatmap(df[numeric_col].corr(), annot = True, cmap = 'YlGnBu')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Independent Features
x = df.drop('Price', axis = 1)
print("------------------< Independent Features >------------------\n")
print(x.columns)

# Dependent Features
y = df['Price']
print("\n------------------< Dependent Features >-------------------\n")
print(y.name)


```

```python
categorical_col = x.select_dtypes(include = ['object']).columns

x_encoded = pd.get_dummies(x, columns = categorical_col)

x_encoded = x_encoded.astype(int)

print(x_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Independent features
X = df.drop("Price", axis=1)

# Apply Standard Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x_encoded)

# Convert back to DataFrame
X_scaled = pd.DataFrame(X_scaled, columns = x_encoded.columns)

# Show first 5 rows
print("=============== First 5 rows After Apply Standard Scaling ============\n")
print(X_scaled.head())
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_encoded, y, test_size = 0.33, random_state = 42)

print("================= Shape ================\n")
print("1) x_train", x_train.shape)
print("\n2) x_test", x_test.shape)
print("\n3) y_train", y_train.shape)
print("\n4) y_test", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train, y_train)

print("<<<<<<<<< Intercept >>>>>>>>>>:\n\n", model.intercept_)
print("\n<<<<<<<<<< Coefficients >>>>>>>>>>>:\n\n", model.coef_)
```

```python
y_pred = model.predict(x_test)

print("================= Predicted Prices ================\n")
print(y_pred[:10])
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)

print("================= R2 Score ================\n")
print("R2 score:", r2)
```

```python
import joblib

joblib.dump(model, "LR_house_price.pkl")
joblib.dump(scaler, "house_scaler.pkl")
joblib.dump(X_scaled.columns.tolist(), "house-columns.pkl")

print("Model saved successfully!")
```
