Replace invalid 0 values with mean

```python
# Replace invalid 0 values with mean
cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

for col in cols:
    mean_value = df[df[col] != 0][col].mean()
    df[col] = df[col].replace(0, round(mean_value, 2))

df.head()
```

Histogram for Numerical Columns

```python
num_cols = ["Pregnancies", "Glucose", "BloodPressure"]

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=20)
    plt.title(f"{col} Distribution")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

Diabetes Outcome Countplot

```python
sns.countplot(x="Outcome", data=df)
plt.title("Diabetes Outcome Distribution")
plt.xlabel("Outcome")
plt.ylabel("Count")
plt.show()
```

Scatter Plot

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Glucose", y="BMI", hue="Outcome", data=df)
plt.title("Glucose vs BMI")
plt.show()
```

Correlation Heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap="Reds")
plt.title("Correlation Heatmap")
plt.show()
```
