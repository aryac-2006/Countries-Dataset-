1) Bar Chart - Gender Distribution

```python
df["Gender"].value_counts().plot(kind = "bar",  color = ["pink","skyblue"])
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

2) Histogram - Age Distribution

```python
plt.hist(df["Age"], bins = 10, color = "lightgreen")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.hist(df["Annual Income (k$)"], bins = 10, color = "lightpink")
plt.title("Annual Income Distribution")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Frequency")
plt.show()
```

3) Scatter Plot - Annual Income VS Spending Score

```python
sns.scatterplot(x = "Annual Income (k$)",
                y = "Spending Score (1-100)",
                hue = "Gender", data = df)

plt.title("Annual Income VS Spending Score")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

4) Count Plot - Spending Score Groups

```python
df["Spending Group"] = pd.cut(df["Spending Score (1-100)"],
                              bins = [0, 40, 80, 100],
                              labels = ["Low", "Medium", "High"])

sns.countplot(x = "Spending Group", hue = "Gender", data = df)
plt.title("Spending Group Distribution")
plt.xlabel("Spending Group")
plt.ylabel("Count")
plt.show( )
```

5) Heatmap - Correlation Matrix

```python
numeric_df = df.select_dtypes(include = ["number"])
sns.heatmap(numeric_df.corr(), annot = True, cmap = "Reds")
plt.title("Correlation Matrix")
plt.show()
```
