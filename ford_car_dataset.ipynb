Q6. (Correlation Heatmap)

Create a heatmap of the correlation matrix for numeric features.                                                                  
Use seaborn with proper annotations.                                            
  Identify which features are highly correlated with the target variable (price).    
  Write your observations in comment

```python
# select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# check correlation with price
corr_price = correlation_matrix['price'].sort_values(ascending=False)
print("\n-------------Correlation with Price-----------\n\n", corr_price)

# Findings:
# 1. The heatmap shows the relationship between all numeric features.
# 2. Values close to +1 indicate a strong positive correlation.
# 3. Values close to -1 indicate a strong negative correlation.
# 4. Features with higher correlation values with price have a greater influence on car price.
# 5. Features with values close to 0 have little or no relationship with price.
```

```python
# Complete preprocessing pipeline for the Ford Car Dataset

# Step 1: Load the dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("ford_car_dataset.csv")

# Step 2: Check missing values
print("1] -------< Missing values in each column >------\n")
print(df.isnull().sum())

# Step 3: Remove duplicate rows
df = df.drop_duplicates()

# Step 4: Histograms
print("\n2] ------------------------< Histogram >---------------------\n")
columns = ["price", "mileage", "year", "engineSize", "mpg"]

plt.figure(figsize=(8, 7))
for i, col in enumerate(columns, 1):
    plt.subplot(3, 2, i)
    plt.hist(df[col])
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# Step 5: Count Plots
print("\n3] ------------------------< Count Plots >---------------------\n")
cat_cols = ["model", "transmission", "fuelType"]

plt.figure(figsize=(15,8))
for i, col in enumerate(cat_cols, 1):
    plt.subplot(2,2,i)
    sns.countplot(data=df, x=col)
    plt.title(col)
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Step 6: Correlation Heatmap
print("\n4] ------------------------< Correlation Heatmap >---------------------\n")
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Step 7: Identify Input and Output Features
X = df.drop("price", axis=1)
y = df["price"]

print("\n5] Input Features:")
print(X.columns)

print("\n6] Output Feature:")
print(y.name)

# Step 8: One-Hot Encoding
df_encoded = pd.get_dummies(df,
                            columns=["model", "transmission", "fuelType"],
                            dtype=int)

# Step 9: Standard Scaling
scaler = StandardScaler()

numeric_cols = ["year", "mileage", "tax", "mpg", "engineSize"]

df_encoded[numeric_cols] = scaler.fit_transform(df_encoded[numeric_cols])

print("\n7] Preprocessing Completed Successfully!")
print(df_encoded.head())
```
